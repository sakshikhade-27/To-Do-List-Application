To-Do List Manager

```python

tasks = []  
status = [] 

def add_task():
    task = input("Enter new task: ")
    tasks.append(task)
    status.append("Pending")
    print("Task added successfully!\n")

def view_tasks():
    if len(tasks) == 0:
        print("No tasks available.\n")
    else:
        print("\nYour Tasks:")
        for i in range(len(tasks)):
            print(i + 1, ".", tasks[i], "-", status[i])
        print()

def update_task():
    view_tasks()
    num = int(input("Enter task number to update status: "))
    if status[num - 1] == "Pending":
        status[num - 1] = "Completed"
    else:
        status[num - 1] = "Pending"
    print("Task status updated!\n")

def delete_task():
    view_tasks()
    num = int(input("Enter task number to delete: "))
    tasks.pop(num - 1)
    status.pop(num - 1)
    print("Task deleted successfully!\n")

while True:
    print("----- To-Do List Manager -----")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Update Task Status")
    print("4. Delete Task")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_task()
    elif choice == "2":
        view_tasks()
    elif choice == "3":
        update_task()
    elif choice == "4":
        delete_task()
    elif choice == "5":
        print("Thank you! Goodbye ðŸ‘‹")
        break
    else:
        print("Invalid choice. Try again.\n")
```
